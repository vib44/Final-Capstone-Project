### We need to clean this dataframe as well
Here, we drop all other columns except 'Districts' and 'Quarters' 

```python
quarters.drop(quarters.iloc[:,3:6],axis=1,inplace=True)
quarters.drop('Quartiers(Quarters)',axis=1,inplace=True)
quarters.columns=['Districts','Quarters']
quarters
```

```python
quarters['Districts']=quarters['Districts'].str.split(" ").str[0]
quarters
```

```python
#We convert dataframes into lists so that the manipulkations become easier
n=quarters['Quarters'].values.tolist()
pc=quarters['Districts'].values.tolist()
pcd=pc[1:]
v=[]      #List to save new QUARTER values
p=[]      #List to save Borough No. values without repitition
c=''      #To save the string to be input in list v
k=''      #To save the extra quarter names that are to be added 
r=1       
d=1
t=0
z=''      
for P,PC in enumerate(pc):       #loop to compare the Borough No.
    if((P+t)<80):      #There are total 80 elements in pc
        c=n[P+t]        #Quarter name
        z=pc[P+t]       #Borough Name
        k=''
        d=1
        p.append(z)     #appending Borough no. to list p
             
        for i,g in enumerate(pcd):          #loop to compare the Borough No.
            if(z==g):                               
                if(c!=n[i+r]):               #To make sure we dont compare the same Quarter with itself
                    k=k+','+n[i+r]
                    d=d+1
                    t=t+1       
            else:
                break
        pcd=pcd[d:]                #As the comparision is done we keep on decreasing the list to eliminate the elements that have been compared
        r=r+d
        c=c+k               #Final string of Neighbourhood/Quarter names
        v.append(c)         #Appending the string in list v
    else:
        break            
    
```

### The Dataframe can now be made :-

```python
dd={'Borough No.':p,'Quarters':v}
ddd=pd.DataFrame(dd)
ddd.head(12)
```

```python
import json
import requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
results = requests.get(url).json()  #making request to the Foursquare API
results
```

```python
# function that extracts the category of the venue
#In the above cell we see the key 'categories' is key to the name of the venue category hence we use row['categories']
#or row['venue.categories'] in case categories key exist inside any dictionary with key as 'venue'
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
```

```python
venues = results['response']['groups'][0]['items'] #obtaining the part that contains information about venues
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]  #adding column names to the dataframe

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)   #obtaining categories for respective venues with the help of get_category_type

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
print('{} venues were returned by Foursquare.'.format(nearby_venues.shape[0]))
nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):  #loop to find nearby values and their information for every Neighborhood
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    return(nearby_venues)
```

```python
Paris_venues = getNearbyVenues(names=info['Name'],
                               latitudes=info['Latitude'],
                                   longitudes=info['Longitude']
                                  )
```

### We use one hot encoding to deal with categorical data while clustering using k means

```python
# one hot encoding
Paris_onehot = pd.get_dummies(Paris_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Paris_onehot['Neighborhood'] = Paris_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Paris_onehot.columns[-1]] + list(Paris_onehot.columns[:-1])
Paris_onehot = Paris_onehot[fixed_columns]

Paris_onehot.head()
```

```python
# Here we're grouping the venue categories with respect to neighbourhood
Paris_grouped =Paris_onehot.groupby('Neighborhood').mean().reset_index()
Paris_grouped
```

## Find top 5 popular venues for each Neighbourhood

```python
num_top_venues = 5

for hood in Paris_grouped['Neighborhood']:         #for loop to find the venue and the frequency of occurence of the venue in a particular Neighbourhood
    print("----"+hood+"----")
    temp = Paris_grouped[Paris_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):                   #function for finding most common venues
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]    #returns the top ten venues with highest frequency of occurrence in a Neighborhood
```

## Find top 10 common venues for each Neighbourhood and put it into a dataframe

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = Paris_grouped['Neighborhood']

for ind in np.arange(Paris_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Paris_grouped.iloc[ind, :], num_top_venues)   #Putting values to each row
neighborhoods_venues_sorted.head()
```

### We take number of clusters as 5


```python
# set number of clusters
kclusters = 5

Paris_grouped_clustering = Paris_grouped.drop('Neighborhood', 1)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Paris_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

## Visualize clusters on the map of Paris using folium

```python
# create map
%matplotlib inline
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster,pc in zip(Paris_merged['Latitude'], Paris_merged['Longitude'], Paris_merged['Name'], Paris_merged['Cluster Labels'],Paris_merged['Postal Codes']):
    label = folium.Popup(str(poi)+','+str(pc) + ' ,Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
